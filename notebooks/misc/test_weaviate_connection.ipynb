```python
import weaviate
from weaviate.client import WeaviateClient
```

```python
weaviate.connect_to_local()
```

```python
connection_params = weaviate.connect.ConnectionParams(
        http={
            "host": "localhost",
            "port": 8080,
            "secure": False
        },
        grpc={
            "host": "localhost",
            "port": 50051,
            "secure": False
        }
    )
```

```python
client = weaviate.WeaviateClient(connection_params)
```

```python
class WeaviateConnectionManager:

    def __init__(self, client: WeaviateClient):
        self.client = client

    def __enter__(self) -> WeaviateClient:
        self.client.connect()
        return self.client

    def __exit__(self, exc_type, exc_val, exc_tb):
        self.client.close()

    
```

```python
with WeaviateConnectionManager(client=client) as weaviate_client:
    print(weaviate_client.collections.list_all())
    
```
